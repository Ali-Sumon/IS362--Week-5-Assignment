```python

# read the csv into a dataframe
weather = pd.read_csv("/Users/aliahsansumon/Documents/IS362/Week5/weather.csv")
```

```python
# Display the columns in the dataframe, and get a peak at how it's formatted
weather.head()
```

```python
weather[((weather.year == 2013) & \
         (weather.month == 2)) & \
        (weather.day == 12) & \
        (weather.wind_speed < 350)] \
        .sort_values('wind_speed', ascending = False)
```

```python
# To account for the fact that the wind data is taken every hour I will take the average hourly 
# windspeed for each airport that day.
weather[((weather.year == 2013) & \
         (weather.month == 2)) & \
        (weather.day == 12) & \
        (weather.wind_speed < 350)] \
        .groupby('origin').wind_speed.mean()
```
